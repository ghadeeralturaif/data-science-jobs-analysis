```r
#The top five the job titles
  DS_job %>%
  filter(!(is.na(job_title))) %>%
  group_by(job_title) %>%
  summarize(count=n()) %>%
  arrange(-count) %>%
  top_n(10) %>%
  mutate(job_title= fct_reorder(job_title, count)) %>%
  ggplot(aes(x=job_title, y=count, fill=count)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  labs( title="The top ten jobe title")
```

```r
#Average salary in USD group by location of company
DS_job %>%
  group_by(company_location) %>%
  summarise(Average_salary = mean(salary_in_usd),
            .groups = "drop") %>%
              top_n(10) %>%
  ggplot(aes(company_location, Average_salary, fill = company_location)) +
  geom_col(position = "dodge") +
  labs(title = "The average salary in USD group by location of company")
```

```r
#the most popular job
DS_job %>%
  group_by(job_title) %>%
  summarize(count = n()) %>%
  arrange(desc(count)) %>%
  top_n(5)
```

```r
#the highest Salary for each level of experience
DS_job %>%
  group_by(experience_level) %>%
  summarise(max_salary = max(salary_in_usd)) %>%
  arrange(desc(max_salary))
```

```r
# the highest Salary of Data Analyst job
DS_job %>%
  filter(job_title %in% "Data Analyst") %>%
  select(c("job_title", "salary_in_usd")) %>%
  arrange(desc(salary_in_usd)) %>%
  top_n(3)
```

```r
# the most ten job title based on company_size
DS_job %>%
  group_by(job_title, company_size) %>%
  summarise(count = n(),
            .groups = "drop") %>%
    top_n(10) %>%
  ggplot(aes(count, job_title, fill = company_size)) +
  geom_col(position = "dodge")
```

```r
# The average salary in USD by employment types
ggplot(data=DS_job) +
geom_bar(aes(x = employment_type, y = salary_in_usd, fill = employment_type), stat = "summary", fun = "mean") +
  labs(title="The average salary in USD by employment types")
```

```r
#the max, median and avarege salary by year
DS_job %>%
  group_by(work_year)%>%
  summarize(
   max_salary=max(salary_in_usd),
   min_salary=min(salary_in_usd),
   median_salary = median(salary_in_usd),
   average_salary=mean(salary_in_usd))
```

```r
#visualization of max, median and avarege salary by year
DS_job %>%
  group_by(work_year)%>%
  summarize(
   median_salary = median(salary_in_usd),
   max_salary=max(salary_in_usd),
   average_salary=mean(salary_in_usd))%>%
  ggplot(mapping = aes(x=work_year, y =salary_in_usd))+
  scale_y_continuous()+
  geom_line(aes(work_year,median_salary , colour = "median_salary"), lwd=1.5)+
  geom_line(aes(work_year, max_salary, colour = "max_salary"), lwd=1.5)+
  geom_line(aes(work_year, average_salary, colour = "average_salary"), lwd=1.5)+
  theme_bw()+
  scale_colour_manual("",
                      breaks = c("median_salary","max_salary", "average_salary"),
                      values = c("darkgoldenrod3", "cadetblue", "chocolate4"))+
  labs(title = "Maximum, median and average salary in Data related jobs in last four years")
```

```r
# Create separate data frames for each year
ds_2020 <- DS_job %>% filter(work_year == 2020)
ds_2021 <- DS_job %>% filter(work_year == 2021)
ds_2022 <- DS_job %>% filter(work_year == 2022)
ds_2023 <- DS_job %>% filter(work_year == 2023)

# count the most job title in each year
ds_summary_2020 <- ds_2020 %>%
  group_by(job_title) %>%
  summarize(count=n()) %>%
  top_n(4)
ds_summary_2021 <- ds_2021 %>%
  group_by(job_title) %>%
  summarize(count=n()) %>%
  top_n(4)
ds_summary_2022 <- ds_2022 %>%
  group_by(job_title) %>%
  summarize(count=n()) %>%
  top_n(4)
ds_summary_2023 <- ds_2023 %>%
  group_by(job_title) %>%
  summarize(count=n()) %>%
  top_n(4)

# Combine the summary data frames and add a column for the year
ds_summary_all <- bind_rows(
  mutate(ds_summary_2020, year = 2020),
  mutate(ds_summary_2021, year = 2021),
  mutate(ds_summary_2022, year = 2022),
  mutate(ds_summary_2023, year = 2023))
```

```r
#the average salary for each job title in each year
ggplot(ds_summary_all, aes(x = count, y = job_title , fill = factor(year))) +
  geom_bar(stat = "identity", position = "dodge") +
  scale_fill_manual(values = c("#e41a1c", "#377eb8", "#4daf4a", "darkgoldenrod3")) +
  labs(title = "The most jobs title by year") +
  theme_bw()
```
